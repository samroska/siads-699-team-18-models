```python
# Import Libraries
import pandas as pd
import altair as alt
```

```python
# Directories
IMAGES_DIR = './Images/BCN20000-all-images-original/bcn_20k_train/'
META_DATA_DIR = './Metadata/'
MODELS_DIR = './Models/'
TEST_IMAGES_DIR = './Images/Images_for_tests/'
TRAINING_RESULTS_DIR = './Training_Results/'
```

```python
# Read dataset for Mobile Images
df = pd.read_csv(META_DATA_DIR + 'PAD-UFES-20-Metadata-One-Hot.csv')
```

```python
# Count number of samples per class
counted_df = df['diagnostic'].value_counts().reset_index()
counted_df.columns = ['Diagnotics', 'Count']
```

```python
# Rename Classes for Chart
counted_df.loc[counted_df['Diagnotics'] == 'ACK','Diagnotics'] = 'actinic keratosis'
counted_df.loc[counted_df['Diagnotics'] == 'BCC','Diagnotics'] = 'basal cell carcinoma'
counted_df.loc[counted_df['Diagnotics'] == 'NEV','Diagnotics'] = 'nevus'
counted_df.loc[counted_df['Diagnotics'] == 'SCC','Diagnotics'] = 'squamous cell carcinoma'
counted_df.loc[counted_df['Diagnotics'] == 'SEK','Diagnotics'] = 'seborrheic keratosis'
counted_df.loc[counted_df['Diagnotics'] == 'MEL','Diagnotics'] = 'melanoma'
counted_df.loc[counted_df['Diagnotics'] == 'MPX','Diagnotics'] = 'monkeypox'
```

```python
# Build Bar Chart
chart = alt.Chart(counted_df).encode(
    x=alt.X(
        'Diagnotics',
        title=None,
        sort='-y',
        axis=alt.Axis(labels=False)
    ),

    y=alt.Y('Count', title='Number of Samples'),
    tooltip=['Diagnotics', 'Count'],

    color=alt.Color(
        'Diagnotics',
        scale=alt.Scale(scheme='category20'),
        legend=alt.Legend(
            titleFontSize=12,
            labelFontSize=12
        )
    )
).properties(
    title='Smartphone Images Dataset: Distribution of Samples',
    width=500,
    height=300
)

bars = chart.mark_bar()

text = chart.mark_text(
    align='center',
    baseline='bottom',
    dy=-5,
    color='black'
).encode(
    text=alt.Text('Count'), color=alt.value('black')
)

bars + text
```
