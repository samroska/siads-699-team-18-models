```python
import zipfile
import tensorflow as tf
from PIL import Image
import os
import shutil
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.utils import load_img
```

```python
# Directories
ORIGINAL_IMAGES_DIR = './Images/PAD-UFES-20-all-images/'
RGB_IMAGES_DIR = './Images/PAD-UFES-20-all-images-RGB/'
```

```python
# Convert images from RGBA to RGB (only used once)
source_dir = ORIGINAL_IMAGES_DIR
destination_dir = RGB_IMAGES_DIR

list_of_files = os.listdir(source_dir)

for filename in list_of_files:
  print(filename)
  filename_with_path = source_dir + filename
  image_rgba = Image.open(filename_with_path)
  print(img_to_array(image_rgba).shape)
  image_rgb = Image.open(filename_with_path).convert('RGB')
  print(img_to_array(image_rgb).shape)
  destination_file_name = destination_dir + filename[0:-4] + '.jpg'
  print(destination_file_name)
  tf.keras.utils.save_img(destination_file_name, image_rgb)
```
